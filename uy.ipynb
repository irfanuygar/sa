```python
print("""*************************************************
            ATM uygulamasına hoş geldiniz
**************************************************""")



adım="irfan"

şifrem="987654"

soyadı="uygar"

girişhakkınız =5
while True:

    kulanıcıadı=input("Kulanıcı Adınızı Giriniz:")
    

    parola=input("Parolanızı Giriniz:")
    

    soyadı=input("soyadınızı giriniz:")
    
    
    if(kulanıcıadı==adım and şifrem!=parola):
        print("Parolanız Yanlış")
        girişhakkınız -=1
        print("Toplam {} Girişhakkınız Kaldı".format(girişhakkınız))
        
        continue

        
    elif(adım!=kulanıcıadı and şifrem==parola):
        print("Kulanıcı Gdınız Yanlış")
        girişhakkınız-=1
        print("Toplam {} Girişhakkınız Kaldı".format(girişhakkınız))
        
        continue

        
    elif(adım!=kulanıcıadı and şifrem!=parola):
        print("Kulanıcı Adınız Ve Parolanız Yanliş")
        girişhakkınız-=1
        print("toplam {} girişhakkınız kaldı".format(girişhakkınız))
        
        continue

        
    elif(adım!=kulanıcıadı and şifrem!=parola):
        print("parolanız ve kullancı adınız yanlış")
        girişhakkınız-=1
        print("toplam {} girişhakkınız kaldı".format(girişhakkınız))
             
       
        
    else:
        print("Giriş Başarılı")
        

    if(girişhakkınız==0):
        break



    bakiye=10000

    while True:

        işlem = input("İşlem Seçiniz :")

        if (işlem == "Çıkışyap"):
            print("YİNE BEKLERİZ...")
            break
        elif (işlem=="1"):
            print("Toplam Bakiyeniz : {} TL dir".format(bakiye))

        elif(işlem=="2"):
            miktar=int(input("Çekmek İstediğiniz Bakiyeyi Giriniz:"))
            if (bakiye - miktar < 0):
                print("yetersiz bakiye")
                print("toplam bakiyeniz : {} TL dir".format(bakiye))
            bakiye -=miktar

        elif(işlem=="3"):
            miktar=int(input("Yatımak İstediğiniz Bakiyeyi Giriniz:"))
            bakiye+=miktar
    
        else:
            print("Geçersiz İşlem")
```
